### Using lists and dictionaries to describe how to efficiently describe instructions

```python
config = [
    {
        'line_instruction': [
            {'character': '-', 'repeat': 10},
            {'character': 'X', 'repeat': 30},
            {'character': '-', 'repeat': 10},
            {'character': '\n', 'repeat': 1},
        ],
        'repeat': 3
    },
    {
        'line_instruction': [
            {'character': '-', 'repeat': 20},
            {'character': 'X', 'repeat': 10},
            {'character': '-', 'repeat': 20},
            {'character': '\n', 'repeat': 1},
        ],
        'repeat': 12
    }
]

output = ''
for section in config:
    section_output = ''
    for instruction in section['line_instruction']:
        section_output += instruction['character'] * instruction['repeat']
    section_output *= section['repeat']
    output += section_output

print(output)
```

### Example 2 -- slightly different way of writing the code to write the T

Two steps
1) Construct all the lines first and then
2) Combine all the lines

```python
t_config = [
    {
        'line_instruction': [
            {'character': '-', 'repeat': 10},
            {'character': 'X', 'repeat': 30},
            {'character': '-', 'repeat': 10}
        ],
        'repeat': 3
    },
    {
        'line_instruction': [
            {'character': '-', 'repeat': 20},
            {'character': 'X', 'repeat': 10},
            {'character': '-', 'repeat': 20}
        ],
        'repeat': 12
    }
]

# 1. create the lines
lines = []
for section in t_config:
    line = ''
    for instruction in section['line_instruction']:
        line += instruction['character'] * instruction['repeat']
    
    for ctr in range(section['repeat']):
        lines.append(line)
    
# 2. combine and print the lines
output = ''
for line in lines:
    output += line + '\n'

print(output)
```
